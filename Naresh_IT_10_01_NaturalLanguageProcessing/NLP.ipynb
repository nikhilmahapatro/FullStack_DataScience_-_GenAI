## Unigrams, Bigrams Trigrams, Ngrams

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
string = "The quick brown fox jumps over the lazy dog near the quiet river bank."
string
```

```python
quotes_tokens = word_tokenize(string)
quotes_tokens
```

```python
len(quotes_tokens)
```

```python
# BIGRAMS

quotes_bigrams = list(nltk.bigrams(quotes_tokens))
quotes_bigrams
```

```python
# TRIGRAMS

quotes_trigrams = list(nltk.trigrams(quotes_tokens))
quotes_trigrams
```

```python
# NGRAMS

quotes_ngrams = list(nltk.ngrams(quotes_tokens, 5))                 # no of word in token = 5
quotes_ngrams
```

## STEMMING

### PorterStemmer

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
pst.stem('affection')
```

```python
pst.stem('playing')
```

```python
pst.stem('maximum effort')
```

```python
words_to_stem = ['giving','given','give','gave']

for words in words_to_stem:
    print(words + ": " + pst.stem(words))
```

```python
words_to_stem = ['giving','eaten','nullified','criticized','fed','crushed','pulling','loving']

for words in words_to_stem:
    print(words + ": " + pst.stem(words))
```

### LancasterStemmer

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
```

```python
words_to_stem = ['giving','eaten','nullified','criticized','fed','crushed','pulling','loving']

for words in words_to_stem:
    print(words + ": " + lst.stem(words))
```

### SnowballStemmer

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')
```

```python
words_to_stem = ['giving','eaten','nullified','criticized','fed','crushed','pulling','loving']

for words in words_to_stem:
    print(words + ": " + sbst.stem(words))
```
